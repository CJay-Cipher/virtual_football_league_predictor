```python
# CONVERT HTML TO DOCX
# pip install docxtpl

from docx import Document
from bs4 import BeautifulSoup

def convert_html_to_docx(html_path, docx_path):
    try:
        doc = Document()

        # Load the HTML content
        with open(html_path, "r", encoding="utf-8") as html_file:
            html_content = html_file.read()

        # Parse the HTML content
        soup = BeautifulSoup(html_content, "html.parser")

        # Extract text and formatting from HTML
        for element in soup.recursiveChildGenerator():
            if element.name == "p":
                paragraph = doc.add_paragraph(element.text)
                if element.get("style"):
                    paragraph.style = element["style"]

        # Save the document as DOCX
        doc.save(docx_path)

        print("HTML converted to DOCX successfully!")
    except Exception as e:
        print("An error occurred:", str(e))

# Specify the paths for the input HTML and output DOCX files
html_file = "output.html"
docx_file = "new.docx"

convert_html_to_docx(html_file, docx_file)
```

```python
import fitz

def replace_text_in_pdf(input_path, output_path, old_text, new_text):
    doc = fitz.open(input_path)

    for page in doc:
        text_instances = page.searchFor(old_text)
        for text_instance in text_instances:
            page.add_redact_annot(text_instance)
            page.apply_redactions()

        page_text = page.get_text()
        modified_text = page_text.replace(old_text, new_text)
        page.set_text(modified_text)

    doc.save(output_path)
    doc.close()

# Usage example
replace_text_in_pdf('sample_cv.pdf', 'output.pdf', 'Ingenious', 'Brilliant')
```

```python
from docx import Document
from pdf2docx import Converter
import tempfile
import os

def convert_pdf_to_doc(pdf_path):
    # Create a temporary file for the output DOCX
    temp_dir = tempfile.mkdtemp()
    temp_file = os.path.join(temp_dir, 'output.docx')

    # Convert PDF to DOCX
    cv = Converter(pdf_path)
    cv.convert(temp_file)
    cv.close()

    return temp_file


def text_cleaner(text):
    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text)
    # Remove newline characters
    text = text.replace('\n', '')
    # Strip leading and trailing white spaces
    text = text.strip()

    return text


def docx_modifier(file_path, old_section, new_section):
    doc = Document(file_path)

    for i, paragraph in enumerate(doc.paragraphs):
        if old_section in text_cleaner(paragraph.text):
            doc.paragraphs[i].text = new_section

    return doc
    

def main():
    pdf_file = "family_consent.pdf"
    output_path = "new.docx"

    file_path = convert_pdf_to_doc(pdf_file)
    
    old_section = text_cleaner(input("Paste the text you want to update here:\n"))
    new_section = """This is an updated sample text"""
    
    updated_docx = docx_modifier(file_path, old_section, new_section)
    updated_docx.save(output_path)

if __name__ == "__main__":
    main()
```
